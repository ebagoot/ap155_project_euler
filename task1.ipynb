# Problem 1
The Collatz sequence is defined as follows $\forall c_n$ $$c_{n+1} = \frac{1}{2}c_n \quad \forall c_n \text{ even} $$ and $$c_{n+1} = 3c_{n} + 1 \quad \forall c_n \text{ odd}$$

```python
#   Initialization of x:integers from 1 to 10,000 <arbitrary limit> and y:their respective T(N) arrays
limit = 1e4
x = np.arange(1, limit+1, step=1, dtype=int)
y = np.zeros(int(limit), dtype=int)
for i in x:
    y[i] = collatz_unity(x[i]) 

#y = np.array(collatz_unity(x))
last_number = 1

def collatz_next(current):          #   Propagates the sequence to the next number. In: current number -> Out: next number
    if (current % 2 == 0):
        current /= 2
    else:
        current = 3*current + 1
    return current

def collatz_unity(c_1):             #   Goes through the whole sequence until reaching the last number. In: start of sequence -> Out: index of last number in sequence.
    index_count = 0
    c_2 = c_1
    while (c_2 != last_number):
        c_2 = collatz_next(c_2)
        index_count += 1
    return index_count

N = 200013  
print(collatz_unity(N))
print(x[3])
```

We then have a function that can find the index of the last number in the sequence before termination (last number = 1).
