# Problem 1
The Collatz sequence is defined as follows, where $\forall \ c_n$: 
$$c_{n+1} = \frac{1}{2}c_n \quad \forall \ c_n \text{ even}$$
and 
$$c_{n+1} = 3c_{n} + 1 \quad \forall \ c_n \text{ odd}$$

The start of this sequence is an arbitrary positive number $N$, and a function $T(N)$ takes the value of the index of the first appearance of $1$ in that sequence.

We can start with creating functions that can generate the next number in a sequence given the above conditions.
With it, we can create an addtional function that facilitates generation of the whole sequence of the starting seed.
  - Ultimately, the design of the problem forgoes having to collect the whole sequence. This can be added in the future.
  - This function will have a loop that continuously generates the next number in the sequence until the designated *"last number"* is generated in the sequence for the first time.
    - We can add the functionality of finding the next $n^{th}$ appearance of the *last number* in the sequence in the future.
